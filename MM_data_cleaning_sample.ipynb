```python
import kagglehub
import os
import pandas as pd

# ── 1. Download dataset ──────────────────────────────────────────────────────
path = kagglehub.dataset_download("metmuseum/the-metropolitan-museum-of-art-open-access")
print("Path to dataset files:", path)
```

```python
# ── 2. List all files in the dataset ────────────────────────────────────────
print("\n── Files in dataset ──")
for f in os.listdir(path):
    full = os.path.join(path, f)
    size_mb = os.path.getsize(full) / (1024 ** 2)
    print(f"  {f}  ({size_mb:.2f} MB)")
```

```python
# ── 3. Load the main CSV ─────────────────────────────────────────────────────
csv_path = os.path.join(path, "MetObjects.csv")
df = pd.read_csv(csv_path, low_memory=False)
```

```python
# ── 4. Basic overview ────────────────────────────────────────────────────────
print(f"\n── Shape: {df.shape[0]:,} rows × {df.shape[1]} columns ──")
print("\n── Column names ──")
print(df.columns.tolist())

print("\n── First 5 rows ──")
display(df.head())

print("\n── Data types & null counts ──")
info = pd.DataFrame({
    "dtype":    df.dtypes,
    "nulls":    df.isna().sum(),
    "null_%":   (df.isna().mean() * 100).round(1),
    
})

# Arrange nulls by descending order
info = info.sort_values(by='nulls', ascending=False)

display(info)
```

```python
# ── 5. Quick stats on key columns ────────────────────────────────────────────
print("\n── Top 10 Departments ──")
display(df["Department"].value_counts().head(10))

print("\n── Is Public Domain? ──")
display(df["Is Public Domain"].value_counts())

print("\n── Objects with images (has Object URL) ──")
has_url = df["Link Resource"].notna().sum()
print(f"  {has_url:,} of {len(df):,} objects have a URL ({has_url/len(df)*100:.1f}%)")
```
